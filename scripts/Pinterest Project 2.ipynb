##Import libraries

```python
import requests
import codecs
import re
from time import sleep
import pandas as pd
import csv
from random import randint
import random
```

##Import text file with board urls and create list

```python
textfile = 'trhealth.txt'

results = []
with open('trhealth.txt') as inputfile:
    for line in inputfile:
        results.append(line.strip().split(','))
        
board_urls = []

for i in range(0,len(results)):
    board_urls.append(results[i][0])
```

```python
print len(board_urls)
board_urls[300:350]
```

##Take a random sample of 100,000 boards

```python
board_urls_sample = random.sample(board_urls,2700)
```

```python
board_urls_sample1 = board_urls_sample
myfile = open('sample_2700boards.csv', 'wb')
wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
wr.writerow(board_urls_sample1)
```

##Limit to only boards named health*

```python
health_sample = []

for url in board_urls_sample:
    words = url.split('-')
    if len(words) == 1:
        word = words[0].split('/')
        if word[-2][-6:-1] == 'healt':
            health_sample.append(url) 
        
```

```python
print len(health_sample)

health_sample
```

##Scrape the boards

```python
board_html = []
formturk = []
number = 1 
boardid = 1

for url in health_sample:
    try:
        r = requests.get(url)
        thetext = r.text
        #thetext = thetext.encode('utf8')
        board_html_dict = {}
        try:
            #board_html_dict['text'] = thetext
            board_html_dict['pinner'] = re.findall('https://www.pinterest.com/(.*?)/', url, re.DOTALL)[0]
            board_html_dict['pinner_name'] = re.findall('<h4 class="fullname">(.*?)</h4>', thetext, re.DOTALL)[0]
            board_html_dict['board'] = re.findall('https://www.pinterest.com/(.*?)/(.*?)/', url, re.DOTALL)[0][1]
            board_html_dict['pinnerinfo'] = re.findall('pinsAndFollowerCount ">(.*?)</div>', thetext, re.DOTALL)[0]
            pininfo = re.findall('<a href="/pin/(.*?)/"(.*?)img        src="https://(.*?)"', thetext, re.DOTALL)
            board_html_dict['pin_no'] = []
            board_html_dict['img_urls'] = []
            board_html_dict['url'] = url
            board_html_dict['board_id'] = boardid
            boardid += 1
            for i in range(0,len(pininfo)):
                board_html_dict['pin_no'].append(pininfo[i][0])
                board_html_dict['img_urls'].append('https://' + str(pininfo[i][2]))            
            board_html_dict['numberofpins'] = len(board_html_dict['img_urls'])
            try: 
                avatar = re.findall('<img src="https://s-media-cache-ak0.pinimg.com/avatars/(.*?)"', thetext, re.DOTALL)
                board_html_dict['pinner_avatar'] = 'https://s-media-cache-ak0.pinimg.com/avatars/' + avatar[0]
            except:
                board_html_dict['pinner_avatar'] = 'NA'
            board_html.append(board_html_dict)
            print number
            number += 1
        except:
            print 'I got nothing.'
    except:
        print 'cant access' + str(url)
    
    try:
        if len(board_html_dict['img_urls']) >= 5:
            formturk_dict = {}
            formturk_dict['url'] = url
            formturk_dict['image_urls'] = board_html_dict['img_urls'][0:5]
            formturk_dict['pinner_avatar'] = board_html_dict['pinner_avatar']
            formturk_dict['pin_no'] = board_html_dict['pin_no']
            formturk_dict['board_id'] = board_html_dict['board_id']
            formturk.append(formturk_dict)
    except:
        print 'No.'
    sleep(random.uniform(.1, .5))
```

```python
formturk400 = formturk[0:400]
```

```python
smallsample = []
for i in range(0, len(formturk400)):
    for k in range(0, len(formturk400[i]['image_urls'])):
        row = {}
        row['url'] = formturk400[i]['url']
        row['pin_url'] = formturk400[i]['image_urls'][k]
        row['pinner_avatar'] = formturk400[i]['pinner_avatar']
        row['board_id'] = formturk400[i]['board_id']
        row['pin_no'] = formturk400[i]['pin_no'][k]
        smallsample.append(row)
```
