##Import libraries

```python
import requests
import codecs
import re
from time import sleep
import pandas as pd
import csv
from random import randint
import random
```

```python
df = pd.read_csv('sample_2700boards.csv')
```

```python
dfList = df['board'].tolist()
len(dfList)
```

```python
board_html = []
number = 1 
boardid = 1

for url in dfList:
    try:
        r = requests.get(url)
        thetext = r.text
        #thetext = thetext.encode('utf8')
        board_html_dict = {}
        try:
            #board_html_dict['text'] = thetext
            board_html_dict['pinner'] = re.findall('https://www.pinterest.com/(.*?)/', url, re.DOTALL)[0]
            board_html_dict['pinner_name'] = re.findall('<h4 class="fullname">(.*?)</h4>', thetext, re.DOTALL)[0]
            board_html_dict['board'] = re.findall('https://www.pinterest.com/(.*?)/(.*?)/', url, re.DOTALL)[0][1]
            board_html_dict['pinnerinfo'] = re.findall('pinsAndFollowerCount ">(.*?)</div>', thetext, re.DOTALL)[0]
            pininfo = re.findall('<a href="/pin/(.*?)/"(.*?)img        src="https://(.*?)"', thetext, re.DOTALL)
            board_html_dict['pin_no'] = []
            board_html_dict['img_urls'] = []
            board_html_dict['url'] = url
            board_html_dict['board_id'] = boardid
            boardid += 1
            for i in range(0,len(pininfo)):
                board_html_dict['pin_no'].append(pininfo[i][0])
                board_html_dict['img_urls'].append('https://' + str(pininfo[i][2]))            
            board_html_dict['numberofpins'] = len(board_html_dict['img_urls'])
            try: 
                avatar = re.findall('<img src="https://s-media-cache-ak0.pinimg.com/avatars/(.*?)"', thetext, re.DOTALL)
                board_html_dict['pinner_avatar'] = 'https://s-media-cache-ak0.pinimg.com/avatars/' + avatar[0]
            except:
                board_html_dict['pinner_avatar'] = 'NA'
            board_html.append(board_html_dict)
            print number
            number += 1
        except:
            print 'I got nothing.'
    except:
        print 'cant access' + str(url)
    
    sleep(random.uniform(.1, .5))
```

```python
board_html
```

```python
sample_2700 = []
for i in range(0, len(board_html)):
    for k in range(0, len(board_html[i]['img_urls'])):
        row = {}
        row['url'] = board_html[i]['url']
        row['pin_url'] = board_html[i]['img_urls'][k]
        row['pinner_avatar'] = board_html[i]['pinner_avatar']
        row['pinner'] = board_html[i]['pinner']
        row['pinner_name'] = board_html[i]['pinner_name']
        row['board'] = board_html[i]['board']
        row['board_id'] = board_html[i]['board_id']
        row['pinnerinfo'] = board_html[i]['pinnerinfo']
        row['pin_no'] = board_html[i]['pin_no'][k]
        row['numberofpins'] = board_html[i]['numberofpins']
        sample_2700.append(row)
```

```python
keys = sample_2700[0].keys()
with open('sample_2700boards_full.csv', 'wb') as output_file:
    dict_writer = csv.DictWriter(output_file, keys)
    dict_writer.writeheader()
    dict_writer.writerows(sample_2700)
```
