```python
avatars_big = []
number = 1

for url in set(boardurls):
    board_lib = {}
    try:
        r = requests.get(url)
        thetext = r.text
        try:
            avatar_urls = re.findall('"image_large_url": "(.*?)"', thetext, re.DOTALL)
            pinner = re.findall('https://www.pinterest.com/(.*?)/', url)[0]
            for i in avatar_urls:
                indexer = i.find(pinner)
                if indexer != -1:
                    board_lib['avatar_url'] = 'https:\/\/s-media-cache-ak0.pinimg.com\/avatars\/' + str(i[indexer:])
                    break    
            board_lib['board_url'] = url
            avatars_big.append(board_lib)
            print number
            number += 1
        except:
            print 'I got nothing for ' + str(url)
    except:
        print 'cant access ' + str(url)
    
    sleep(random.uniform(.1, .5))
```

```python
keys = avatars_big[0].keys()
with open('Avatars_big.csv', 'wb') as output_file:
    dict_writer = csv.DictWriter(output_file, keys)
    dict_writer.writeheader()
    dict_writer.writerows(avatars_big)
```
